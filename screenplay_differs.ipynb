```python
def runFileLinesLen(fname):
    file1 = fname.replace(newRepo,oldRepo).replace("/", "\\")
    file2 = fname.replace("/", "\\")
    d = Differ()
    with open(file1) as f1:
        f1_text = f1.readlines()
    with open(file2) as f2:
        f2_text = f2.readlines()
    skip = 0
    f1fix = []
    f2fix = []
    for f in f1_text:
        if f.strip().startswith('--'):
            skip += 1
        else:
            f1fix.append(f.strip())
    for f in f2_text:
        if f.strip().startswith('--'):
            skip += 1
        else:
            f2fix.append(f.strip())
    return len(f1fix), len(f2fix)

from difflib import Differ  
  
def runFileLinesDif(fname):  
    file1 = fname.replace(newRepo,oldRepo).replace("/", "\\")  
    file2 = fname.replace("/", "\\")  
    d = Differ()  
  
    with open(file1) as f1, open(file2) as f2:  
        f1_text = f1.readlines()  
        f2_text = f2.readlines()  
  
    f1fix = [f.strip() for f in f1_text if not f.strip().startswith('--')]  
    f2fix = [f.strip() for f in f2_text if not f.strip().startswith('--')]  
  
    diff = list(d.compare(f1fix, f2fix))  
    return fname.replace(newRepo,'').replace('custom_scripts\screenplays\\',''), [d for d in diff if d[0] in ('-', '+')]  
```
