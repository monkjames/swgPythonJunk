This is the regex used to extract just the loot groups from a mobile template

```lootGroups\s?=\s?{(?:[^{}]|({(?:[^{}]|({(?:[^{}]|({[^{}]*}))*}))*}))*},```

````python
import os
import pandas as pd
import re
from difflib import Differ
````

````python
dOld = 'C:\\Users\\micha\\source\\swg\\MMOCoreORB\\bin\\scripts\\mobile\\'
dNew = 'C:\\Users\\micha\\Documents\\source\\wun1\\MMOCoreORB\\bin\\scripts\\mobile\\'
errors = 0
````

````python
dfiles = []
dfiles.append(["path", "oldLG", "newLG"])

def runFiles(npath):
    dlist = os.listdir(npath)    
    for d in dlist:
        if os.path.isfile(npath + "\\" + d):
            try:
                oldy, newy = getFileContents(npath + "\\" + d)
                oldLG = extractLootGroups(oldy)
                newLG = extractLootGroups(newy)
                dfiles.append([npath + "\\" + d, oldLG, newLG])
            except:
                print("oops")
        else:
            runFiles(npath + "\\" + d)

def getFileContents(fname):
    oldFile = fname.replace(dNew,dOld).replace("/", "\\")
    newFile = fname.replace("/", "\\")
    with open(oldFile) as f1:
        f1_text = f1.read()
    with open(newFile) as f2:
        f2_text = f2.read()
    return f1_text, f2_text

def extractLootGroups(fText):
    re_occurences = "lootGroups\s?=\s?{(?:[^{}]|({(?:[^{}]|({(?:[^{}]|({[^{}]*}))*}))*}))*},"
    return re.findall(re_occurences,fText,flags=re.IGNORECASE)
````

````python
runFiles(dNew)
````

````python
cdif = []
for f in dfiles:
    if len(f[1]) != 0 or len(f[2]) != 0:
        if f[1] != f[2]:
            cdif.append(f)
````

````python
df = pd.DataFrame(cdif)
````
